```python
#Simple Line Chart
#Below, we have created a simple line chart using samples of apple ohlc dataset. 
#We'll be adding watermarks to this chart.
import pandas as pd

apple_df = pd.read_csv("https://raw.githubusercontent.com/slowjerm1/MatPlotLib_Tutorial/main/data/apple_share_price.csv")
apple_df.columns = apple_df.columns.str.strip()
apple_df = apple_df.set_index('Date')

print(apple_df)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,8))

plt.plot(apple_df.index, apple_df.Close, color="dodgerblue", linewidth=2.0);

plt.yticks(range(160,341,20), [f"{val} $" for val in range(160,341,20)])

plt.xlabel("Date", fontdict=dict(fontsize=16, fontweight="bold"))
plt.ylabel("Close Price ($)", fontdict=dict(fontsize=16, fontweight="bold"))
plt.title("Apple Close Prices 2019-2020", loc="left", pad=10, fontdict=dict(fontsize=20, fontweight="bold"));
```

```python
#Image as Watermark
#In this section, we have explained how to add an image as a watermark to our chart.

#The below code simply loads an image named cc.png in memory using image submodule of matplotlib. It's a small image
```

```python
from matplotlib import image

logo = image.imread("th.jpeg")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,8))

plt.plot(apple_df.index, apple_df.Close, color="dodgerblue", linewidth=2.0);

fig.figimage(logo,425,250, alpha=0.5)

plt.yticks(range(160,341,20), [f"{val} $" for val in range(160,341,20)])

plt.xlabel("Date", fontdict=dict(fontsize=16, fontweight="bold"))
plt.ylabel("Close Price ($)", fontdict=dict(fontsize=16, fontweight="bold"))
plt.title("Apple Close Prices 2019-2020", loc="left", pad=10, fontdict=dict(fontsize=20, fontweight="bold"));
```

```python
#Below, we have simply added a theme named fivethirtyeight to chart. 
#If you want to learn how to add themes to your matplotlib charts then 
#feel free to check our simple tutorial on the concept.

#Style Matplotlib Charts | Change Theme of Matplotlib Charts
```

```python
import matplotlib.pyplot as plt

with plt.style.context("fivethirtyeight"):
    fig = plt.figure(figsize=(15,8))

    plt.plot(apple_df.index, apple_df.Close, color="dodgerblue", linewidth=2.0);

    fig.figimage(logo,425,250, alpha=0.5)

    plt.yticks(range(160,341,20), [f"{val} $" for val in range(160,341,20)])

    plt.xlabel("Date", fontdict=dict(fontsize=16, fontweight="bold"))
    plt.ylabel("Close Price ($)", fontdict=dict(fontsize=16, fontweight="bold"))
    plt.title("Apple Close Prices 2019-2020", loc="left", pad=10, fontdict=dict(fontsize=20, fontweight="bold"));
```
